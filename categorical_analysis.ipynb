```python
import pandas as pd 
```

```python
df = pd.read_csv('googleplaystore.csv')
df.head(10)
```

```python
df['Category']
```

```python
df.dropna(inplace=True)
df
```

```python
df['Category'].unique()
```

```python
data = {'Category' : [], 'App Count' : []} 
popular_category = ""
popularity = 0 
for i in df['Category'].unique(): 
  data['Category'].append(i)  
  data['App Count'].append(len(df[df['Category'] == i ]))
  if popularity < len(df[df['Category'] == i ]):
    popularity = len(df[df['Category'] == i ])
    popular_category = i 
category_count = pd.DataFrame(data)
print(f'Most popular category : {popular_category} with {popularity} apps.')
category_count
```

```python
total_number_apps = df['App'].count()
df['Type'].unique()
```

```python
c = 0 
for i in df['Type']:
  if i == 'Free':
    c += 1
print("Number of Free Apps :", c )
```

```python
paid_apps = total_number_apps -c  
print('Number of Paid Apps :', paid_apps )
```

```python
percentage_paid_apps = paid_apps / total_number_apps 
print('Percentage of paid applications :', round(percentage_paid_apps * 100, 2), "%" )
print('Percentage of Free Applications :',100 - round(percentage_paid_apps*100 , 2), "%" )
```

```python
data_type = { "Type" : [], "App Count" : []}
for i in df['Type'].unique() : 
  data_type['Type'].append(i) 
  data_type['App Count'].append(len(df[df['Type'] == i ]))
type_df = pd.DataFrame(data_type) 
type_df 
```

```python
data_content = { "Content Rating" : [], "App Count" : []}
for i in df['Content Rating'].unique() : 
  data_content['Content Rating'].append(i) 
  data_content['App Count'].append(len(df[df['Content Rating'] == i ]))
type_df = pd.DataFrame(data_content) 
type_df 
```
