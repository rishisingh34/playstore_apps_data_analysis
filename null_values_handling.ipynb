```python
import pandas as pd 
from sklearn.impute import SimpleImputer 
import numpy as np 
```

```python
df = pd.read_csv('googleplaystore.csv') 
df.head(10)
```

```python
df.sample(10)
```

```python
imputer = SimpleImputer(missing_values=np.nan , strategy='mean')
```

```python
imputer.fit(df.iloc[:, 2:3]) 
x = imputer.transform(df.iloc[:, 2:3])
```

```python
df.iloc[:, 2:3] = x 
```

```python
df 
```
